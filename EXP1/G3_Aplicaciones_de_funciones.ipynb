```python
#1.
import matplotlib.pyplot as plt
import numpy as np
from scipy.optimize import fsolve

def h(x):
    return 2*(x-2)**2

def i(x):
    return x+1

x=np.arange(-5, 100, 0.01)

plt.plot(x, h(x),color='m',label='h(x)')

plt.plot(x, i(x),color='g',label='i(x)')

plt.ylim(-1, 10) #determina el recorrido o los valores Y
plt.xlim(-2, 10) #determina el dominio o los valores x

plt.title('Comparacion de las funciones')

plt.grid(True)
plt.legend()
plt.show()
```

```python
#2.
from scipy.optimize import fsolve
import numpy as np

x = np.linspace(0, 15, 1)
solucion_h = fsolve(h, x)
solucion_i = fsolve(i, x)

print(f'La funcion h(x) interseca al eje x en x = {solucion_h[0]:.0f} y la funcion i(x) en x ={solucion_i[0]:.0f}')
```

# Problema 3

Sean las funciones:

\begin{split}
w(x) & = 0,6x^{2} + 1,5\\
t(x) & = 0,8x^{3} - 10x + 10
\end{split}

<br>

Considerando que $\hspace{0.2cm}$ $-3 \leq x \leq 4$, $\hspace{0.2cm}$ $-4 \leq y \leq 24$ $\hspace{0.2cm}$ realice lo siguiente:

1. Grafique ambas funciones en un mismo gráfico, indicando unidad de medida. Utilice *Matplotlib*.
2. Determine el intervalo en que $w(x) \geq t(x)$.
3. Encuentre los puntos en que $t(x) = 6$.


```python
#1
import matplotlib.pyplot as plt
import numpy as np
from scipy.optimize import fsolve

def w(x):
    return 0.6*x**2 + 1.5
def t(x):
    return 0.8*x**3 - 10*x +10

x=np.arange(-10, 10, 0.01)

plt.plot(x, w(x),color='m',label='w(x)')

plt.plot(x, t(x),color='g',label='t(x)')

plt.ylim(-4, 24) #determina el recorrido o los valores Y
plt.xlim(-3, 4) #determina el dominio o los valores x

plt.title('Comparacion de las funciones')

plt.grid(True)
plt.legend()
plt.show()
```

```python
#2.
import numpy as np
from scipy.optimize import fsolve

def w(x):
    return 0.6*x**2 + 1.5
def t(x):
    return 0.8*x**3 - 10*x +10

def inter(x):
    return w(x)-t(x)

x = np.linspace(0, 10, 1)

solucion = np.around(fsolve(inter, x),decimals=2)

interseccion = np.unique(solucion)

print(f'Los puntos de interseccion de las funciones estan en x ={interseccion}')
```

```python
#3

def seis(x):
    return t(x) - 6

x = np.linspace(-4, 4, 4)

solucion = np.around(fsolve(seis, x), decimals= 4)

interseccion = np.unique(solucion)

print(f'La funcion t(x) es igual a 6 en x = {interseccion}')
```
